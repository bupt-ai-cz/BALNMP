add status label to json file

```python
import pandas as pd
import json
```

```python
clinical_data = pd.read_excel('/content/drive/MyDrive/cs230output/patient-clinical-data.xlsx')
# map ALN status 'N0', 'N+(>2)', 'N+(1-2)' to -> 0, 1, 2
clinical_data['status_label'] = clinical_data['ALN status'].map({'N0': 0, 'N+(>2)': 1, 'N+(1-2)': 2})
all_status_label = clinical_data[['Patient ID', 'status_label']]
all_status_label.rename(columns={'Patient ID': 'id'}, inplace=True)
test_json = pd.read_json('/content/drive/MyDrive/cs230output/test-type-0.json').rename(columns={'label': 'metastasis_label'})
train_json = pd.read_json('/content/drive/MyDrive/cs230output/train-type-0.json').rename(columns={'label': 'metastasis_label'})
val_json = pd.read_json('/content/drive/MyDrive/cs230output/val-type-0.json').rename(columns={'label': 'metastasis_label'})
# join on 'id'
test_json = test_json.merge(all_status_label, on='id')
train_json = train_json.merge(all_status_label, on='id')
val_json = val_json.merge(all_status_label, on='id')
# output as new json files
test_json.to_json('/content/drive/MyDrive/cs230output/updated_test-type-0.json', orient='records')
train_json.to_json('/content/drive/MyDrive/cs230output/updated_train-type-0.json', orient='records')
val_json.to_json('/content/drive/MyDrive/cs230output/updated_val-type-0.json', orient='records')
```

checking the output

checking all image data referenced in JSON files exist

```python
# !unzip -q "/content/drive/MyDrive/cs230output/paper_patches.zip" -d "/content/drive/MyDrive/cs230output/images" # unzip the image
```

```python
import os
count = 0
with open('/content/drive/MyDrive/cs230output/updated_train-type-0.json', 'r') as f:
    training_data = json.load(f)
for i in range(len(training_data)):
  patch_paths = training_data[i]['patch_paths']
  for j in range(len(patch_paths)):
      if not os.path.isfile(os.path.join('/content/drive/MyDrive/cs230output/images/', patch_paths[j])):
        count += 1
        print(patch_paths[j])
print('#patches not exist for training data: '+ str(count) )
```

```python
count = 0
with open('/content/drive/MyDrive/cs230output/updated_val-type-0.json', 'r') as f:
    val_data = json.load(f)
for i in range(len(val_data)):
  patch_paths = val_data[i]['patch_paths']
  for j in range(len(patch_paths)):
      if not os.path.isfile(os.path.join('/content/drive/MyDrive/cs230output/images/', patch_paths[j])):
        count += 1
        print(patch_paths[j])
print('#patches not exist for validation data: '+ str(count) )
```
